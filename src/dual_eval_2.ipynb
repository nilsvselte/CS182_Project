```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../src/results/dual_mixed_mean.csv", index_col=0)

b_values = [int(c.split("_")[1]) for c in df.columns]

plt.figure(figsize=(6,4))
for a, row in df.iterrows():
    plt.plot(b_values, row.values, marker="o", label=f"A={a}")
plt.xlabel("Quadratic examples (B)")
plt.ylabel("Mean MSE")
plt.title("Dual evaluation")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```
